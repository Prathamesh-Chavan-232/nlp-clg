```python
from nltk.corpus import stopwords
text = """This is a sample sentence and we are going to remove the stopwords from this""" 
stop_words = set(stopwords.words("english"))
word_tokens = nltk.word_tokenize(text)
removed_stopwords = [word for word in word_tokens if word not in stop_words]
stopwords_found = [word for word in word_tokens if word in stop_words]
len(removed_stopwords)
```
